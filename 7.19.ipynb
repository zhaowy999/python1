##### 循环
- 循环是一种控制语句块重复执行的结构

## while 循环
- 当一个条件保持真的时候while循环重复执行语句
- while 循环一定要有结束条件，否则很容易进入死循环
- while 循环的语法是：

    while loop-contunuation-conndition:

        Statement

## 示例：
sum = 0

i = 1

while i <10:

    sum = sum + i
    i = i + 1

## 错误示例：
sum = 0

i = 1

while i <10:

    sum = sum + i

i = i + 1
- 一旦进入死循环可按 Ctrl + c 停止

# 验证码
- 随机产生四个字母的验证码，如果正确，输出验证码正确。如果错误，产生新的验证码，用户重新输入。
- 验证码只能输入三次，如果三次都错，返回“别爬了，我们小网站没什么好爬的”
- 密码登录，如果三次错误，账号被锁定


```python
import random
x = random.randint(1000,9999)
print(x)
while 1 :
    a = eval(input('请输入四个数字'))
    if a==x:
        print('正确')
        break
        
    
```

```python
import random
x = random.randint(1000,9999)
print(x)
i = 0
while i < 3 :
    a = eval(input('请输入四个数字'))
    if a==x:
        print('正确')
        break
    else:
        print('错误')
        i += 1
    if i==3:
        print('别爬了傻狗')      
```

## for 循环
- Python的for 循环通过一个序列中的每个值来进行迭代
- range(a,b,k)， a，b，k 必须为整数
- a: start
- b: end
- k: step
- 注意for 是循环一切可迭代对象，而不是只能使用range

```python
for i in range(10):    #区间前闭后开
    print(i)
```

```python
for i in range(11,1,-2):
    print(i)
```

```python
a = {1,2,3}
a.__iter__
```

```python
for i in a:
    print(i)
```

## 嵌套循环
- 一个循环可以嵌套另一个循环
- 每次循环外层时，内层循环都会被刷新重新完成循环
- 也就是说，大循环执行一次，小循环会全部执行一次
- 注意：
> - 多层循环非常耗时
  - 最多使用3层循环

## EP：
- 使用多层循环完成9X9乘法表
- 显示50以内所有的素数

```python
for i in range(1,10):
    for j in range(1,i+1):
        print(j,'x',i,'=',i * j,end=' ' )
    print()
```

```python
for i in range(1,10):
    for j in range(i,10):
        print(j,'x',i,'=',i * j,end=' ' )
    print(i,j)
```

```python
for i in range(2,51):
    for j in range(2,i):
        if i%j==0:
            break
    else:
        print(i)
```

```python
for i in range(2,51):
    for j in range(2,i):
        print(i,j)
```

## 关键字 break 和 continue
- break 跳出循环，终止循环
- continue 跳出此次循环，继续执行

```python
for i in range(10):
    if i == 5:
        continue    #不打印5
    print(i)
```
