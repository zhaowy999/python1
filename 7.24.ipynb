## 下角标运算符 []
- 一个字符串是一个字符序列，可以通过索引进行访问
- 观察字符串是否是一个可迭代序列  \__iter__

```python
c = 'JNDNDddd'
print(c)
c[0]      #正索引从零开始
```

```python
c[-1]     #负索引从-1开始
```

```python
for i in  'liuzhizhi':
    print(i)
```

```python
c = 'liuzhizhi'
i = 0
while 1:
    a = c[i]
    i += 1
    print(a)
    if i == 9:
        break
```

## 切片 [start: end]
- start 默认值为0
- end 默认值为-1

#####如果步长为正数那么start <end 
####如果步长为负，那么start > end
####如果想要整体反转（左右），[::_1]

```python
a = '刘支支和赵爹爹'
a[0:3]  #切片 前闭后开
```

```python
a = '刘支支和赵爹爹'
a[-7:-4] 
```

```python
a = '刘支支和赵爹爹'
a[2:0:-1] 
```

```python
a[::]  #a[0:6:1]
```

```python
a[::-1]  #a[6:0:-1]
```
